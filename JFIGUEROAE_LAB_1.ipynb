```
def calculate_area(length, width):
    """ Calculates the area of a rectangle """
    return length * width
l = 5
w = 10
print ('Area: ', calculate_area(l, w))
print ('Length: ', l)
print ('Width: ', w)

def calculate_volume(length, width, depth):
    """ Calculates the volume of a rectangular prism """
    return length * width * depth
d = 3
print ('Volume: ', calculate_volume(l, w, d))
print ('Length: ', l, "width: ", w, "depth: ", d)
```

**Clases y funciones**

```
import yfinance as yf
import numpy as np
import pandas as pd

# This is a plotting library for pretty pictures.
import matplotlib.pyplot as plt

class stock:
    def __init__(self, ticker, open, close, volume):
        self.ticker = ticker
        self.open = open
        self.close = close
        self.volume = volume
        self.rate_return = float(close)/open - 1

    def update(self, open, close):
        self.open = open
        self.close = close
        self.rate_return = float(self.close)/self.open - 1

    def print_return(self):
        print(self.rate_return)

apple = stock('AAPL', 143.69, 144.09, 20109375)
google = stock('GOOG', 898.7, 911.7, 1561616)

apple.ticker
google.print_return()
google.update(912.8,913.4)
google.print_return()
```

```
def has_a_vowel(word):
    """
    Checks to see whether a word contains a vowel
    If it doesn't contain a conventional vowel, it
    will check for the presence of 'y' or 'w'. Does
    not check to see whether those are in the word
    in a vowel context.
    """
    vowel_list = ['a', 'e', 'i', 'o', 'u']

    for vowel in vowel_list:
        if vowel in word:
            return True
    # If there is a vowel in the word, the function returns, preventing anything after this loop from running
    return False

print(has_a_vowel("apple"))  # True, porque contiene 'a' y 'e'
print(has_a_vowel("sky"))    # False, porque no tiene 'a', 'e', 'i', 'o' o 'u' (aunque tiene 'y', no lo verifica en contexto de vocal)


```

```
import numpy as np

# Definir matrices
A = np.array([
    [1, 2, 3, 12, 6],
    [4, 5, 6, 15, 20],
    [7, 8, 9, 10, 10]
])

B = np.array([
    [4, 4, 2],
    [2, 3, 1],
    [6, 5, 8],
    [9, 9, 9]
])

# Verificar dimensiones antes de multiplicar
if A.shape[1] == B.shape[0]:  # La cantidad de columnas de A debe coincidir con las filas de B
    resultado = np.dot(A, B)  # Multiplicación matricial
    print("Resultado de la multiplicación:")
    print(resultado)
else:
    print("Las dimensiones de las matrices no permiten la multiplicación.")
```

```
import matplotlib.pyplot as plt
import yfinance as yf

data_all = yf.download("AAPL,MSFT", start="2021-01-01", end="2022-01-01")
data_all.head()

data = data_all["Close"]
data.head()
data['MSFT'].head()

# Plot a histogram using 20 bins
plt.hist(data['MSFT'], bins=20)
plt.xlabel('Price')
plt.ylabel('Number of Days Observed')
plt.title('Frequency Distribution of MSFT Prices, 2021');

```

```
# Remove the first element because percent change from nothing to something is NaN
R = data['MSFT'].pct_change()[1:]

# Plot a histogram using 20 bins
plt.hist(R, bins=20)
plt.xlabel('Return')
plt.ylabel('Number of Days Observed')
plt.title('Frequency Distribution of MSFT Returns, 2021');
```

```
plt.scatter(data['MSFT'], data['AAPL'])
plt.xlabel('MSFT')
plt.ylabel('AAPL')
plt.title('Daily Prices in 2021');
```

```
R_msft = data['MSFT'].pct_change()[1:]
R_aapl = data['AAPL'].pct_change()[1:]

plt.scatter(R_msft, R_aapl)
plt.xlabel('MSFT')
plt.ylabel('AAPL')
plt.title('Daily Returns in 2021');
```

```
#Line graph
plt.plot(data['MSFT'])
plt.plot(data['AAPL'])
plt.ylabel('Price')
plt.legend(['MSFT', 'AAPL']);

```
